```python
#Global Variables
color = ['Black', 'Red', 'Green', 'Yellow', 'Orange', 'Blue', 'Pink', 'Magenta', 'Grey', 'Purple']

#Graph Sizes
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,5

#Warning disablied because anomaly is caused due to injuries 
import warnings
warnings.filterwarnings('ignore')
```

```python
#Dynamic Function to protray insights
def plotGraphs(x,y):
    for name in Players:
        plt.plot(y[Pdict[name]], c= color[Pdict[name]], ls='--', marker='o', ms=7, label=Players[Pdict[name]]) 
        
    plt.xticks(list(range(len(x))), x, rotation = 'vertical')
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.show()
```

```python
#Free Throw Attempts per game
FTAttemp = FTA / Games
plotGraphs(Seasons, FTAttemp)
```

Domain Knowledge and Insight

Dwight Howard plays the center position as a result, the chances of him getting fouled and getting a FTA is much higher than a guard. This is because as a center you spend more time in the paint with the ball, thus increasing ones odds of getting a FTA. This makes sense because according to the graph, he has got the most FTA during the ten seasons than any other top 10 player. Also, another trend here is that every single player had more FTA in 2005 season then they did in 2014 season. This pattern indicates the transition of the NBA style of play. Today, the game is based off jump shoots and three pointers than driving to the paint and trying to get the foul.

```python
#Accuracy of Free Throws
Accuracy = FT / FTA
plotGraphs(Seasons, Accuracy)
```

Domain Knowledge and Insight

This graph also provides us an explanation on why maybe Dwight Howard had the most FTA over the ten seasons. As you can see, Howard has the lowest FT accuracy which means fouling Howard may have been a tactic for many teams. This can be done to decrease the chances of the other team to get a two-point play or trying get ball possession back. This graph also demonstrates that in order to be a top player in the NBA, one should have a good FT accuracy
